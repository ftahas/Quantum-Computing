```python
from qiskit import QuantumCircuit

# 1 qubit, 1 classical bit
qc = QuantumCircuit(1, 1)

# Apply a Hadamard gate
qc.h(0)

# Measure qubit 0 into classical bit 0
qc.measure(0, 0)

qc
```

```python
from qiskit_aer import Aer
from qiskit import transpile

backend = Aer.get_backend("aer_simulator")

# Transpile for backend
compiled_circuit = transpile(qc, backend)

# Run
job = backend.run(compiled_circuit, shots=1000)
result = job.result()

counts = result.get_counts()
counts
```

```python
import numpy as np
from qiskit import QuantumCircuit

def qft(n: int) -> QuantumCircuit:
    qc = QuantumCircuit(n, name=f"QFT({n})")
    for j in range(n):
        qc.h(j)
        for k in range(j+1, n):
            angle = np.pi / (2 ** (k - j))
            qc.cp(angle, k, j)  # controlled phase
    # Swap to reverse bit order
    for i in range(n//2):
        qc.swap(i, n-1-i)
    return qc

def iqft(n: int) -> QuantumCircuit:
    qc = qft(n).inverse()
    qc.name = f"IQFT({n})"
    return qc
```

```python
qft4 = qft(4)
print(qft4.draw("text"))
# qft4.draw("mpl")
```

```python
from qiskit.circuit import Gate

def cU_2_mod_15(power: int) -> Gate:
    """
    Returns a controlled gate implementing multiplication by 2^(power) mod 15
    on a 4-qubit target register, in a compact toy construction.
    power is 0,1,2,3 corresponding to 2,4,8,1 mod 15 cycles.
    """
    qc = QuantumCircuit(4)

    # These are compact permutations for N=15, a=2.
    # power=0: multiply by 2 mod 15 (on states in the cycle)
    # power=1: multiply by 4 mod 15
    # power=2: multiply by 8 mod 15
    # power=3: multiply by 1 mod 15 (identity)
    #
    # For the standard toy order-finding, we only need the action on the subspace
    # reached from |0001>. These SWAP patterns reproduce that cycle.

    if power == 0:
        qc.swap(0, 1)
        qc.swap(1, 2)
        qc.swap(2, 3)
    elif power == 1:
        qc.swap(0, 2)
        qc.swap(1, 3)
    elif power == 2:
        qc.swap(0, 3)
        qc.swap(1, 2)
    elif power == 3:
        pass
    else:
        raise ValueError("power must be 0..3 for this toy demo")

    U = qc.to_gate(label=f"U(2^{power} mod 15)")
    cU = U.control(1)
    return cU
```

```python
block = cU_2_mod_15(0)
# Put it into a circuit to draw
tmp = QuantumCircuit(1+4)
tmp.append(block, [0,1,2,3,4])
print(tmp.draw("text"))
# tmp.draw("mpl")
```

```python
def shor_period_finding_15_a2(nc: int = 4) -> QuantumCircuit:
    qc = QuantumCircuit(nc + 4, nc)

    # 1) put counting qubits in superposition
    for q in range(nc):
        qc.h(q)

    # 2) target register init to |1>
    qc.x(nc + 0)  # |0001> on the 4 target qubits (little-endian style)

    # 3) controlled-U^{2^k}
    # Apply in increasing k corresponding to qubit q
    for q in range(nc):
        # We only need powers modulo 4 because order r=4 here
        power = q % 4
        qc.append(cU_2_mod_15(power), [q] + list(range(nc, nc+4)))

    # 4) inverse QFT on counting register
    qc.append(iqft(nc), range(nc))

    # 5) measure counting register
    qc.measure(range(nc), range(nc))
    return qc

qc = shor_period_finding_15_a2(4)
print(qc.draw("text"))
# qc.draw("mpl")
```

```python
backend = Aer.get_backend("aer_simulator")
tqc = transpile(qc, backend, optimization_level=1)
result = backend.run(tqc, shots=4096).result()
counts = result.get_counts()
counts
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

```python
from fractions import Fraction

def guess_r_from_bitstring(bitstring: str, nc: int) -> int:
    s = int(bitstring, 2)
    frac = Fraction(s, 2**nc).limit_denominator(15)
    return frac.denominator

# Example: take the most common non-zero outcome
top = sorted(counts.items(), key=lambda x: x[1], reverse=True)
top[:5]
```

```python
for b, c in top[:6]:
    r_guess = guess_r_from_bitstring(b, 4)
    print(b, "count:", c, "-> r_guess:", r_guess)
```

```python
import math

def shor_factors_from_r(N: int, a: int, r: int):
    if r % 2 != 0:
        return None
    x = pow(a, r//2, N)
    if x == N-1:
        return None
    p = math.gcd(x-1, N)
    q = math.gcd(x+1, N)
    if p*q == N and p not in (1, N) and q not in (1, N):
        return (p, q)
    return None

print(shor_factors_from_r(15, 2, 4))
```
